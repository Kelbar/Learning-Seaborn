```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

iris = sns.load_dataset('iris')
iris.head()
```

**PairGrid**

```python
# We will use PairGrid. In simple words is the pairplot method customizable
sns.PairGrid(iris)
```

```python
# We can map over the PairGrid
pair = sns.PairGrid(iris)
pair.map(plt.scatter)
```

```python
# We can also map to upper,lower, and diagonal

g = sns.PairGrid(iris)

g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

**PairPlot**

```python
#pairplot is a simple version of PairGrid
sns.pairplot(iris)
```

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species',palette='magma')
```

**FacePlot**

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
# With FaceGrid we can set a col and a row and map it with a one or two variable. Depends on the plot that we use
sns.FacetGrid(tips, row='smoker', col='time')
```

```python
fg = sns.FacetGrid(tips, row='smoker', col='time')

fg.map(sns.kdeplot, 'total_bill')
```

```python
# If the plot needs two arguments you can past it wit FaceGrid
fg.map(plt.scatter,'total_bill', 'tip')
```
