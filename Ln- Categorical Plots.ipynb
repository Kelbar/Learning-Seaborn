```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

There are a few main plot types for this:

    factorplot
    boxplot
    violinplot
    stripplot
    swarmplot
    barplot
    countplot

```python
sns.barplot(x='smoker', y='tip', data=tips)
```

```python
#barplot can set a estimator such a np.std or your own custom function
sns.barplot(x='smoker', y='tip', data=tips, estimator=np.std)
```

```python
#countplot is similar to count
sns.countplot(x='sex', data=tips)
```

```python
#Boxplot is able to show more informatio, min, Q1, mid, Q2, max and also we can use hue to compare
sns.boxplot(x='smoker',y='total_bill', data=tips)
```

```python
sns.boxplot(x='total_bill',y='smoker', data=tips, orient="h")
```

```python
sns.boxplot(x='smoker',y='total_bill', data=tips, hue = 'sex')
```

```python
#Violin plot is similar to box plot but with more information and a nice way to compare
sns.violinplot(x='day',y='total_bill', data=tips)
```

```python
sns.violinplot(x='day',y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x ='day', y='tip', data=tips, jitter=False, hue='smoker', dodge= True)
```

```python
#Jitter will separate the points and makes it more readable
sns.stripplot(x ='day', y='tip', data=tips, jitter=True, hue='smoker')
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
#We can combine Swarm with Violin
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='white')
```

```python
#We can call any plot we factorplot or catplot using kind
sns.catplot(x='day', y='tip', data=tips, kind='box')
```
