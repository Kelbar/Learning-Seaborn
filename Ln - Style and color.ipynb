```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

***Style***

We can set different styles with sns.set_style (darkgrid, whitegrid, dark, white, ticks)

```python
sns.countplot(x='day', data=tips)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='day', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='day', data=tips)
```

```python
sns.set_style('dark')
sns.countplot(x='day', data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='day', data=tips)
```

**We can also remove the spine**

```python
sns.set_style('ticks')
sns.countplot(x='day', data=tips)
sns.despine()
```

```python
sns.set_style('ticks')
sns.countplot(x='day', data=tips)
sns.despine(left=True, bottom=True)
```

**Sizes and aspects**

We can combine with Matplotlib(plt.figure(figsize=)) to set the size

```python
# Non Grid Plot
plt.figure(figsize=[3,8])
sns.countplot(x='day', data=tips)
```

```python
# Grid Type Plot
sns.lmplot(x='total_bill',y='tip',height=2,aspect=4,data=tips)
```

**Scales and context**

The set_context()(paper, notebook, talk, poster) allows you to override default parameters:

```python
sns.set_context('poster')
sns.countplot(x='smoker', data=tips)
```

```python
sns.set_context('paper')
sns.countplot(x='smoker', data=tips)
```

```python
sns.set_context('poster', font_scale=3)
sns.countplot(x='smoker', data=tips)
```

**Palettes**

There are many option of palettes to select for de color lines
https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
sns.set_context('notebook')
sns.lmplot(x='total_bill',y='tip', data=tips, hue='smoker', palette='Spectral')
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, hue='smoker', palette='hsv')
```
